## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Assuming 'Credit_Score' is categorical, map to numerical values:

credit_score_mapping = {'Poor': 1, 'Standard': 2, 'Good': 3}

df['Credit_Score_Numeric'] = df['Credit_Score'].map(credit_score_mapping)

df
````

````python
# Filtering numerical columns

numerical_df = df.select_dtypes(include=np.number)
numerical_df['Credit_Score_Numeric'] = df['Credit_Score_Numeric']
numerical_df.head()
````

````python
# Calculate the correlation between annual income and credit score:

correlation = df['Annual_Income'].corr(df['Credit_Score_Numeric'])

print(f"Correlation between Annual Income and Credit Score: {correlation}")
````

````python
# Calculate the correlation between credit utilization ratio of the customer and their credit score:

correlation_utilization = df['Credit_Utilization_Ratio'].corr(df['Credit_Score_Numeric'])

print(f"Correlation between Credit Utilization Ratio and Credit Score: {correlation_utilization}")
````

````python
# Calculate the correlation between number of loans taken by the customer and their credit score:

correlation_loans = df['Num_of_Loan'].corr(df['Credit_Score_Numeric'])

print(f"Correlation between Total Number of Loans and Credit Score: {correlation_loans}")
````

````python
# Assuming a credit utilization ratio below 35% is considered as a good credit utilization ratio.
filtered_df = df[df['Credit_Utilization_Ratio'] < 35]
````

````python
# Calculate the correlation between number of credit cards taken by the customer and credit score if the credit utilization is low:

correlation_num_cards = filtered_df['Num_Credit_Card'].corr(filtered_df['Credit_Score_Numeric'])
print(f"Correlation between Number of Credit Cards and Credit Score (with low utilization): {correlation_num_cards}")
````

````python
# Calculate the correlation between number of delayed payments made by the customer and their credit score:

correlation_delayed_payments = df['Num_of_Delayed_Payment'].corr(df['Credit_Score_Numeric'])
print(f"Correlation between Number of Delayed Payments and Credit Score: {correlation_delayed_payments}")
````

````python
# Filtering the data where the credit score is Good

good_credit_score = filtered_df[filtered_df['Credit_Score'] == 'Good']

good_credit_score.head()
````

````python
# Filtering the data where the credit score is Standard

standard_credit_score = filtered_df[filtered_df['Credit_Score'] == 'Standard']

standard_credit_score.head()
````

````python
# Filtering the data where the credit score is Poor

poor_credit_score = filtered_df[filtered_df['Credit_Score'] == 'Poor']

poor_credit_score.head()
````

````python
# Number of cards impact on poor credit score

count_of_customers_poor = poor_credit_score.groupby("Num_Credit_Card")["Num_Credit_Card"].value_counts().sort_values(ascending=False)

count_of_customers_poor
````

````python
# Number of cards impact on standard credit score

count_of_customers_standard = standard_credit_score.groupby("Num_Credit_Card")["Num_Credit_Card"].value_counts().sort_values(ascending=False)

count_of_customers_standard
````

````python
# Number of cards impact on good credit score

count_of_customers_good = good_credit_score.groupby("Num_Credit_Card")["Num_Credit_Card"].value_counts().sort_values(ascending=False)

count_of_customers_good
````

````python
# Binning Age

age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
df['age_group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65, float('inf')], labels=age_labels, include_lowest=True)
````

````python
# Binning credit utilisation ratio

Credit_utilisation_bins = ['20-28', '29-32', '33-36', '37-50']
df['Credit_Utilization_Ratio_Bin'] = pd.cut(df['Credit_Utilization_Ratio'], bins=[0, 28, 32, 36, 50], labels=Credit_utilisation_bins)
````

````python
# Binning number of credit enquiries

credit_enquiries_bins = ['0-3', '3-5', '6-9', '9+']
df['Credit_Enquiries_Bin'] = pd.cut(df['Num_Credit_Inquiries'], bins=[0, 3, 5, 9, float('inf')], labels=credit_enquiries_bins, include_lowest=True)
````

````python
# Binning delay from due data

delay_bins = ['0-10', '11-20', '21-30', '30+']
df['Delay_from_due_date_Bin'] = pd.cut(df['Delay_from_due_date'], bins=[0, 10, 20, 30, float('inf')], labels=delay_bins, include_lowest=True)
````

### What all manipulations have you done and insights you found?



*   Assuming credit score is categorical mapped credit score previously (Good, Standard, poor) to (3, 2, 1) respectively.
*   Filtered the columns for numerical type for plotting heatmap and pair plot to get overview of the data.
*   Checked correlation between Annual Income and Credit Score and found out Annual Income has a positive correlation of 0.21.
*   Checked whether correlation between Credit Utilization Ratio and Credit Score would be higher than the correlation between Number of Loans taken and Credit Score and found out Credit Utilization Ratio has a positive correlation of 0.04 whereas Number of Loans taken has negative correlation of -0.35.
*   Checked whether correlation between Number of Credit Cards and Credit Score if the Credit Utilization is low and found out Number of Credit Cards has a negative correlation of -0.40.
*   Checked the correlation between Number of Delayed Payments and Annual Income and found out Number of Delayed Payments has a negative correlation of -0.37.
*   Filtered data with respect to credit score for Good, Standard and Poor respectively to know how the spread of number of Credit Cards across all credit scores.
*   Using pd.cut() method to bin the continuous numerical variables for better visualization based on the data.
